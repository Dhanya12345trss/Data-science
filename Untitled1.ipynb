```python
# for Visualization

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('Universities.csv')
```

```python
df1
```

```python
plt.hist(df1['SAT'])
plt.xlabel('SAT SCORES')
plt.ylabel('FREQUENCY')
plt.title('Histogram of SAT SCORES')
plt.show()
```

```python
sns.histplot(df1['SAT'])
# show is used to hide that starting info
plt.show()
```

```python
sns.histplot(df1['SAT'],color='lightpink')
plt.xlabel('SAT SCORES')
plt.ylabel('FREQUENCY')
plt.title('Histogram of SAT SCORES')
plt.show()
```

```python
# go = g stands for color, o stands for circle
#in this i won't give any 2nd column to compare so it with the index
plt.plot(df1['SAT'],'go')
```

```python
plt.plot(df1['SAT'],df1['Top10'],'yo')
plt.xlabel('SAT SCORES')
plt.ylabel('Top10')
plt.title('scatteer plot of SAT SCORES vs Top10')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Top10'],'rs')
plt.xlabel('SAT SCORES')
plt.ylabel('Top10')
plt.title('scatteer plot of SAT SCORES vs Top10')
plt.show()
```

```python
#histogram + scatterplot
sns.pairplot(df1)
```
